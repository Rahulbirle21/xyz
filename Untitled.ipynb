```python
## From the missing table we can recognize two groups with a pattern, group 1 ('Promo2SinceWeek','Promo2SinceYear','PromoInterval') and second group ('CompetitionOpenSinceMonth','CompetitionOpenSinceYear'). Columns in each group have the same number of missing values. From the column names and the data fields' description we can understand that the values of the columns in each group share a common information. Let's investigate.
```

```python
# dataframe containing the missing rows for columns in group 1
group1_df = store_df[store_df['Promo2SinceWeek'].isna()]
group1_df.head(10)
```

```python
## We can see that the Promo2 column is 0 for the first 10 rows. From intution we can infer that if there is no promotion, the values in the columns in gropu 1(Promo2SinceWeek, Promo2SinceYear, and PromoInterval) should be null. Let's prove it.
```

```python
group1_df[group1_df['Promo2'] == 0][['Promo2']].value_counts()
```

```python
group1_df[['Promo2SinceWeek', 'Promo2SinceYear','PromoInterval']].isnull().sum()
```

```python
## Yes, we are right. All rows with 0 value in their Promo2 columns have null values in the columns Promo2SinceWeek, Promo2SinceYear, and PromoInterval.Now let's impute the missing values in the columns Promo2SinceWeek and Promo2SinceYear with 0 to convey the meaning of absence of any promotion since year and week.
```

```python
def fix_missing_value(df, cols, value):
    for col in cols:
        count = df[col].isna().sum()
        df[col] = df[col].fillna(value)
        if type(value) == 'str':
            print(f"{count} missing values in the column {col} have been replaced by \'{value}\'.")
        else:
            print(f"{count} missing values in the column {col} have been replaced by {value}.")
```

```python
fix_missing_value(store_df, ['Promo2SinceWeek', 'Promo2SinceYear'], 0)
```

```python
## For the column PromoInterval, we will impute it with '0,0,0,0' as the other values follow list-type format containing four months. Let's check.
```

```python
# all the unique values 
store_df['PromoInterval'].value_counts()
```

```python
fix_missing_value(store_df, ['PromoInterval'], '0,0,0,0')
```

```python
## Now it's time for the second group (i.e CompetitionOpenSinceMonth and CompetitionOpenSinceYear).
```

```python
## From the column names and the data fields' description in the challange document we can understand that if CompetitionOpenSinceMonth and CompetitionOpenSinceYear are null it means there was already a competition when the store was opened. So we need impute the missing values in a meaningful way. Before doing that we will create a new column named 'CompetitionBeforeStoreOpened' so that we can make use of this information in our analysis and prediction. This column will hold values of 0 and 1, where 0 indicates the absence of competition at the time the store was opened and 1 indicates the presence of competition at the time the store was opened.
```

```python
# deriving new column
store_df['CompetitionBeforeStoreOpened'] = store_df['CompetitionOpenSinceYear'].apply(lambda x: 1 if np.isnan(x) else 0)
```

```python
## Now, we will impute missing values in the column CompetitionOpenSinceYear and CompetitionOpenSinceMonth with minimum year value and minimum month value respectively. This kind of imputing assumes that the stores with missing values in these columns were opened from the beginning itself.
```

```python
# Identifying minimum year value given in the dataset
min_year = store_df['CompetitionOpenSinceYear'].min()
min_year
```

```python
# Identifying minimum month value given in the dataset
min_month = store_df['CompetitionOpenSinceMonth'].min()
min_month
```

```python
# impute the column CompetitionOpenSinceYear with 1900 and the column CompetitionOpenSinceMonth with 1
fix_missing_value(store_df, ['CompetitionOpenSinceYear'], min_year)
fix_missing_value(store_df, ['CompetitionOpenSinceMonth'], min_month)
```

```python
## Now we are left with missing values in the column 'CompetitionDistance' which has only three missing values. For limited number of missing values, we can assume that the  competitions for these stores are very far away and have almost no impact on these stores that they weren't measured when collecting the data. For this reason, we will impute this values with the maximum competition distance.
```

```python
fix_missing_value(store_df, ['CompetitionDistance'], store_df['CompetitionDistance'].max())
```

```python
# Creating a function to convert columns with float dtype into integer dtype
def convert_to_int(df, columns):
    for col in columns:
        df[col] = df[col].astype("int64")
```

```python
# Creating a function to check mixed dtype in the data
def show_cols_mixed_dtypes(df):
    mixed_dtypes = {'Column': [], 'Data type': []}
    for col in df.columns:
        dtype = pd.api.types.infer_dtype(df[col])
        if dtype.startswith("mixed"):
            mixed_dtypes['Column'].append(col)
            mixed_dtypes['Data type'].append(dtype)
    if len(mixed_dtypes['Column']) == 0:
        print('None of the columns contain mixed types.')
    else:
        print(pd.DataFrame(mixed_dtypes))
```

```python
# Identifying columns with mixed datatypes in store_df
show_cols_mixed_dtypes(store_df)
```

```python
# Creating new features in train_df : 'Year','Month','DayOfMonth','WeekOfYear','weekday' to extract better insights from the data.
train_df['Year'] = train_df['Date'].apply(lambda x: x.year)
train_df['Month'] = train_df['Date'].apply(lambda x: x.month)
train_df['DayOfMonth'] = train_df['Date'].apply(lambda x: x.day)
train_df['WeekOfYear'] = train_df['Date'].apply(lambda x: x.weekofyear)
train_df['weekday'] = train_df['DayOfWeek'].apply(lambda x: 0 if (x in [6, 7]) else 1)
```

```python
train_df.sample(5)
```

```python
# creating a new function to build new features from 'PromoInterval' column in store_df
def getMonth(month_list, index):
    months = ['0', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
    month_list = month_list.split(',')
    month = month_list[index]
    return months.index(month)
```

```python
# split the PromoInterval column into 4 columns
store_df['PromoInterval0'] = store_df.PromoInterval.apply((lambda x: getMonth(x, 0)))
store_df['PromoInterval1'] = store_df.PromoInterval.apply((lambda x: getMonth(x, 1)))
store_df['PromoInterval2'] = store_df.PromoInterval.apply((lambda x: getMonth(x, 2)))
store_df['PromoInterval3'] = store_df.PromoInterval.apply((lambda x: getMonth(x, 3)))
```

### Sales before, after and during Christmas

```python
before_index = (df["Date"] >= pd.to_datetime("2013-11-25")) & (df["Date"] < pd.to_datetime("2013-12-25"))
before_xmass = df[before_index].groupby("Date").agg({"Sales": "mean"})

during_index = (df["Date"] >= pd.to_datetime("2013-12-25")) & (df["Date"] < pd.to_datetime("2013-12-30"))
during_xmass = df[during_index].groupby("Date").agg({"Sales": "mean"})

after_index = (df["Date"] >= pd.to_datetime("2014-01-01")) & (df["Date"] < pd.to_datetime("2014-02-02"))
after_xmass = df[after_index].groupby("Date").agg({"Sales": "mean"})


plt.figure(figsize=(13, 6))

sns.lineplot(x = before_xmass.index, y = before_xmass["Sales"], label='Before Christmas')
sns.lineplot(x = during_xmass.index, y = during_xmass["Sales"], label='During Christmas')
sns.lineplot(x = after_xmass.index, y = after_xmass["Sales"], label='After Christmas')

plt.title("\n Christmass Sales 2013/2014 \n", size=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel(xlabel="Date", fontsize=16)
plt.ylabel(ylabel="Sales", fontsize=16)
plt.show()
```

# Conclusion of EDA

* The most selling and crowded store type is b.
* The most selling and crowded assortment is b.
* Sales is highly correlated to customers.
* For all stores, promotion leads to increase in Sales and Customers. But promotions have low impact on store type b and assortment b when comparing to the other store types and assortments. So promotions should be applied more in the other store types and assortments.
* Store type b is the most opened store type on all weekdays and has its highest sales on sundays than the other days.
* Assortment a and b are the most opened assortments on all weekdays and have their highest sales on sundays than the other days. The number of cusomers increase dramatically on sundays for assortment a while it is similar with other week days for assortment b.
* Sales are increased during Chirstmas week, especially the week before. This might be due to the fact that people buy more beauty products or some common medicines for precaution during Christmas celebration.
* Stores with the smallest competition distance have the highest sales. This indicates that the stores are located at city center or near hospitals. Even though having large distance between competitors is ideally considered great for sales, stores located at city centers will still get more sales regardless of competitors distance as there are more customers at city centers.
